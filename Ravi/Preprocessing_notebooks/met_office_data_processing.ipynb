```
# Retrieving the precipitation data
path = '/content/drive/My Drive/3YP/data/met_office/'    # when running in a notebook in colab

# put the name of the file
filename = "output_5b952037fce4693e254d31e4faf8d392_20201229_001955/subset_2020-12-28T23-33-08_HadREM3-GA705-r001i1p02335 - subset_2020-12-28T23-33-08_HadREM3-GA705-r001i1p02335.csv"
#filename = "output_1b6c9fc5bae406f4c6f4449009b53880_20201229_001449/subset_2020-12-28T23-27-18_HadREM3-GA705-r001i1p02335 - subset_2020-12-28T23-27-18_HadREM3-GA705-r001i1p02335.csv" # the name of the csv file
df = pd.read_csv(path+filename, error_bad_lines = False)               # converting the csv file to a dataframe
```

```
# now we only want to extract the rows with the averages
row_start = 14                # enter the row number on which the data start
new_df = df[row_start::4]
```

```
# now we only want to extract certain columns
df_columns = new_df[['header length', 'Unnamed: 3']]

# convert the data types
df_columns['header length'] = pd.to_datetime(df_columns['header length'])
df_columns['Unnamed: 3'] = pd.to_numeric(df_columns['Unnamed: 3'])

df_columns.columns = ['Date', 'Mean Air Tempeature']
#df_columns.columns = ['Date', 'Precipitation Rate']
```
