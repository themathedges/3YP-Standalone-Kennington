### Modules, Packages and Libraries

Python's way of being 'lazy but smart'

**Module** is a file which contains various Python functions and global variables. It is simply just .py extension file which has python executable code.

**Package** is a simple directory structure having a collection of modules. It must contains an \__init__\.py file as a flag so that the python interpreter processes it as such.

**Library** is a collection of packages or code for distribution which provides the functionality to perform many (often related) tasks without writing your own code.

# Q2 ANSWER 
2. daily maximum half-hourly power over June and December

````
# a starting point... 
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("https://bitbucket.org/scotwheeler/leo_3yp_2020/raw/1977824169f7551bccaf3bd1d7f734728e1a69e3/data/example_PV_data.csv",
                 index_col=0, parse_dates=True, dayfirst=True)

july = df.loc['2019-07-01 00:30:00':'2019-07-31 23:30:00']
dec = df.loc['2019-12-01 00:30:00':'2019-12-31 23:30:00']

daily_max_energy_july = [] #initalise empty list 
daily_max_energy_dec = []

for i in range(0,len(july),48): # iterate through every half-hour in one day, then skip to the next day 
    # find the maximum half hourly energy in a day 
    # 
    # iloc[] locates the 48 slots in one day then we find the max energy value out of those 48 which is the max energy of that day. We append this max daily value to array called 'daily_max_july'
    daily_max_energy_july.append(july.iloc[i:i+47].max()) 
    # to convert the energy to power we *2 to each element. This is cos the energy is for 1/2 hr time periods and 'power = energy*time'
    daily_max_power_july = [element*2 for element in daily_max_energy_july]

for j in range(0,len(dec),48): # iterate through every half-hour in one day, then skip to the next day

    daily_max_energy_dec.append(dec.iloc[j:j+47].max()) # find the maximum half hourly energy in a day
    daily_max_power_dec = [element*2 for element in daily_max_energy_dec]

fig, axs = plt.subplots(2, sharex=True) # common x-axis for both
fig.suptitle('Daily Maximum Half-hourly Power for July and December')
axs[0].plot(daily_max_power_july)
axs[0].set_title('Daily Maximum Power for July')
axs[1].plot(daily_max_power_dec, 'tab:orange')
axs[1].set_title('Daily Maximum Power for December')
axs[1].set_xlabel('Days in Month')
axs[0].set_ylabel('Max Power (kWh)')
axs[1].set_ylabel('Max Power (kWh)')

````
