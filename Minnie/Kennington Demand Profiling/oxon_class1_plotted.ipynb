# Plot Scot's Original Load Data

In this notebook I will plot out the original load data scot had in the model he gave us. So i can see what sort of seasonal variation it had and see if it was taking weekends vs weekday differences into account.

```python
# importing the modules
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import pickle
import datetime
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
```

```python
# load in scot's load data 

df = pd.read_csv('../../data/oxon_class1_year_load.csv')

print(df.info())
print(df.head())
```

```python
load = df['power kW'].values

print(load)
```

```python
daily_mean = []

for i in range(365):
    #daily_mean(i) = mean(ken_avg_nd_load_profile[i*48:48*(i+1)])
    daily_mean = np.append(daily_mean,np.mean(load[i*48:48*(i+1)])) 
    
print(len(daily_mean))
#print(daily_mean)
```

```python
fig2, ax2 = plt.subplots(figsize=(20,10))
day = np.arange(1,366)
ax2.plot(day,daily_mean)
ax2.set(xlabel='Days of year', ylabel='Power (kW)',title="Domestic Demand Data from Scot's code")

ax2.xaxis.label.set_size(20)
ax2.yaxis.label.set_size(20)
ax2.title.set_size(20)

ax2.tick_params(axis='both', which='major', labelsize=15)
```
